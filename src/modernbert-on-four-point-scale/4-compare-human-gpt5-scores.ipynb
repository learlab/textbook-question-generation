```python
import pandas as pd

df1 = pd.read_csv("../../data/itell-question-data-w-gpt5-scores.csv")
df1 = df1.rename(columns={"score": "ensemble_score"})
df2 = pd.read_csv("../../data/cri_annotations.csv")
df2 = df2.rename(columns={"score": "human_score"})
df = pd.merge(
    df1, df2, left_on="id", right_on="question_id", how="left", suffixes=["", "_remove"]
)
```

```python
df = df.drop(
    columns=[
        "candidate",
        "labels",
        "question_id",
        "reference",
        "text",
        "id",
        "condition",
        "chunk_header_remove",
        "question_remove",
    ]
)
```

```python
train_data = df[df["human_score"].isna()]
test_data = df[~df["human_score"].isna()]
test_data
```

```python
# train_data.to_csv("../../data/authentic_train_data_v2.csv", index=False)
test_data.to_csv("../../data/authentic_test_data_v2.csv", index=False)
```
