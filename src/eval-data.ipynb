```python
from supabase import create_client, Client
import pandas as pd
import json
```

```python
supabase: Client = create_client(URL, KEY)
```

```python
response = supabase.table('automatic-short-answer-scoring').select("*").execute()
```

```python
df = pd.DataFrame.from_dict(response.data, orient='columns').iloc[31:]
```

```python
df.correct_response = df['correct_response'].apply(lambda x: 'correct_answer' if x == 'Yes' else 'incorrect_answer')
```

```python
df.to_csv('results.csv')
df
```
